```python
import pandas as pd
from pull_PW_data import pull_data
```

```python
sql_query = """select
    l.*
    , b.id as building_id
    , b.display_address
    , b.city
    , b.state
    , b.zip
from listings l
inner join buildings b
    on l.building_id = b.id 
inner join building_geographies bg
    on bg.building_id = b.id 
inner join geographies g
    on bg.geography_id = g.id
inner join listing_histories lh
    on lh.listing_id = l.id 
    and COALESCE(lh.update_transaction, lh.created_at) = (
    SELECT max(COALESCE(lh2.update_transaction, lh2.created_at)) 
    FROM listing_histories lh2 
    WHERE lh2.listing_id = l.id)
where g.name = 'Greater Cincinnati'
and lh.update_transaction >= '2022-10-01' and lh.update_transaction <= '2023-10-01'
and lh.status_code = 500
and b.in_search = true;"""
```

```python
df_out = pull_data(sql_query, 'data/cincy_all_listings.csv')
```
