```python
import pandas as pd 
import pickle as pkl
import random

from tqdm.notebook import tqdm
```

```python
cols = []
no_media = 0
all_media = []
this_year = []
for i in tqdm(range(9)):
    with open(f'data/normalized_icaar_df_{i}.pkl', 'rb') as f:
        tmp_df = pkl.load(f)
    for d in tmp_df['data']:
        cols += list(d.keys())

    for index, row in tmp_df.iterrows():
        try:
            if row['data']['L_InputDate'][:4] == '2023':
                this_year += row['data']['Media']
        except KeyError:
            continue
        try:
            all_media += row['data']['Media']
        except KeyError:
            no_media += 1

cols = set(cols)
len(cols)   
```

```python
cols
```

```python
no_media
```

```python
len(all_media)
```

```python
len(this_year)
```

```python
cols = {}
for m in all_media:
    for c in m.keys():
        if m[c]:
            try:
                cols[c] += 1
            except KeyError:
                cols[c] = 1
```

```python
cols_ty = {}
for m in this_year:
    for c in m.keys():
        if m[c]:
            try:
                cols_ty[c] += 1
            except KeyError:
                cols_ty[c] = 1
```

```python
cols
```

```python
random.shuffle(all_media)
sample_values = []
for c in cols:
    values = ''
    for m in all_media:
        if len(values.split('|')) == 6:
            break
        try:
            if m[c]:
                values += m[c] + '|'
        except KeyError:
            continue
    sample_values.append(values[:-1])

retval = {'column_name': cols.keys(),
          'num_not_null': cols.values(),
          'sample_values': sample_values,
          'total': [len(all_media)]*len(cols.keys())}
pd.DataFrame(retval).to_csv('data/media_dist.csv', index=False)
```

```python
random.shuffle(this_year)
sample_values = []
for c in cols_ty:
    values = ''
    for m in this_year:
        if len(values.split('|')) == 6:
            break
        try:
            if m[c]:
                values += m[c] + '|'
        except KeyError:
            continue
    sample_values.append(values[:-1])
    
retval = {'column_name': cols_ty.keys(),
          'num_not_null': cols_ty.values(),
          'sample_values': sample_values,
          'total': [len(this_year)]*len(cols_ty.keys())}
pd.DataFrame(retval).to_csv('data/media_dist_2023.csv', index=False)
```
