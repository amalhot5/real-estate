```python
import pandas as pd
```

```python
pw = pd.read_csv('PW_Listings_Change_Log_2024_03_21 (1).csv')
rls = pd.read_csv('PLS_Listings_with_Expiration_2024_03_20.csv')
```

```python
pw['prev_expiration_date'] = [x.split(',')[0][2:-1] for x in pw['expiration_date_changes']]
pw['next_expiration_date'] = [x.split(',')[1][2:-2] for x in pw['expiration_date_changes']]
```

```python
pw.head()
```

```python
ret_df = pw.merge(rls,how='inner', left_on='mls_id', right_on='listing_key').drop(['listing_id', 'listing_key'], axis='columns')
```

```python
ret_df.columns
```

```python
ret_df.columns = ['mls_id', 'address', 'unit_number', 'sale_or_rental', 'status',
       'last_price', 'listing_created_date', 'expiration_date_changes',
       'change_date', 'current_expiration_date', 'first_agent_name',
       'first_agent_email', 'first_agent_phone', 'second_agent_name',
       'second_agent_email', 'second_agent_phone', 'third_agent_name',
       'third_agent_email', 'third_agent_phone', 'brokerage',
       'prev_expiration_date', 'next_expiration_date', 'rls_expiration_date',
       'rls_list_price']
```

```python
ret_df = ret_df[['mls_id', 'address', 'unit_number', 'sale_or_rental',
       'status', 'last_price', 'listing_created_date', 'change_date',
       'prev_expiration_date', 'next_expiration_date', 'current_expiration_date',
       'brokerage','rls_expiration_date', 'rls_list_price', 
       'first_agent_name', 'first_agent_email', 'first_agent_phone',
       'second_agent_name', 'second_agent_email', 'second_agent_phone',
       'third_agent_name', 'third_agent_email', 'third_agent_phone',]]
```

```python
ret_df.head()
```

```python
ret_df.to_csv("Sotheby's Expiration Date Changes.csv", index=False)
```
