```python
def get_listing_id(id):
    if pd.isna(id):
        return 0
    try:
        return int(id)
    except:
        return int(id.split('-')[1])
```

```python
listings['rebny_id'] = [get_listing_id(x) for x in listings['rebny_id']]
```
