```python
import pandas as pd
import json
```

```python
df = pd.read_csv('CRMLS_Euphrates_agents_without_offices_2024_05_01.csv')
def load_json(s):
    try:
        return json.loads(s.replace("'", '"'))
    except:
        try:
            return json.loads(s.replace("DON'T", 'DONOT').replace('""', '"').replace("'",'"'))
        except:
            return None
df['data'] = [load_json(x) for x in df['data']]
```

```python
df = pd.read_csv('[External] CRMLS Members without Offices - Active Agents with No Office Info.csv')
def load_json(s):
    try:
        return json.loads(s.replace("'", '"'))
    except:
        try:
            return json.loads(s.replace("DON'T", 'DONOT').replace('""', '"').replace("'",'"'))
        except:
            return None
df['data'] = [load_json(x) for x in df['data']]
```

```python
df.head()
```

```python
df['data'] = [pd.DataFrame(x, index=[y]) for x,y in zip(df['data'], range(len(df)))]
df.head()
```

```python
df['data'][0]
```

```python
all_data = pd.concat(df['data'].values)
```

```python
all_data.head()
```

```python
df.merge(all_data, how='inner', left_index=True, right_index=True).drop(columns='data')
```
