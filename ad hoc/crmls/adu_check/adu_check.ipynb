```python
df = pd.read_csv('adu_check.csv')
df.columns = ['source_id', 'ADU_json']
df.head()
```

```python
has_seller=[]
for adu, id in zip(df['ADU_json'], df['source_id']):
    tmp = ast.literal_eval(adu)
    for i in tmp:
        if 'ADUKitchenFeatures' in i.keys():
            if 'Seller' in i['ADUKitchenFeatures']:
                has_seller.append(id)
                print(i['ADUKitchenFeatures'])
```
