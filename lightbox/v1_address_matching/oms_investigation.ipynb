```python
def standardize_address(address: str) -> str:
    address = address.upper()
    address = address.replace('(', '')
    address = address.replace(')', '')
    for index, row in suffixes.iterrows():
        if ' ' + row['Street Suffix'] in address:
            address = address.replace(row['Street Suffix'], row['Abbreviation'])
    return address
```

```python
off_market_sales = off_market_sales[~off_market_sales['DEED_LID'].isin(second_match['DEED_LID'])]
third_match = pw_closings[~(pw_closings['listing_id'].isin(first_match['listing_id'])) & ~(pw_closings['listing_id'].isin(second_match['listing_id']))]
len(third_match)
```

```python
retlist = []
for a, VAL_TRANSFER, DATE_TRANSFER in tqdm(zip(off_market_sales['SITE_ADDR'], off_market_sales['VAL_TRANSFER'], off_market_sales['DATE_TRANSFER']), total=len(off_market_sales)):
    found = False
    for p_address, l_id, close_price, close_date in zip(third_match['normalized_address'], third_match['listing_id'], third_match['close_price'], third_match['close_date']):
        if (a in p_address or p_address in a) and (abs((close_price - VAL_TRANSFER)/close_price) <= 0.3) and (abs(close_date - DATE_TRANSFER) <= timedelta(days=90)):
            retlist.append(l_id)
            found = True
            break
    if not found:
        retlist.append(None)
off_market_sales['matched_listing_id'] = retlist
```

```python
def fuzzy_match(lb_address: str, pw_address: str, listing_id, lb_price, pw_price, lb_date, pw_date): #, lb_zip, pw_zip):
    l = lb_address.split()
    p = pw_address.split()
    m_count = 0
    if len(l) > len(p):
        len_address = len(l)
        for i in l:
            if i in p: m_count += 1
    else:
        len_address = len(p)
        for i in p:
            if i in l: m_count += 1
    if m_count / len_address >= 0.66 and (abs((lb_price - pw_price) / pw_price) <= 0.3) and (abs(pw_date - lb_date) <= timedelta(days=90)):# and lb_zip == pw_zip:
        return listing_id
    else:
        return None
```

```python
retlist = []
for lb_address, lb_price, lb_date, lb_zip in tqdm(zip(off_market_sales['SITE_ADDR'], off_market_sales['VAL_TRANSFER'], off_market_sales['DATE_TRANSFER'], off_market_sales['SITE_ZIP']), total=len(off_market_sales)):
    check = third_match[(third_match['postal_code'].str.contains(str(lb_zip))) | (third_match['postal_code'] == lb_zip)]
    found = False
    for pw_address, pw_price, pw_date, l_id in zip(check['normalized_address'], check['close_price'], check['close_date'], check['listing_id']):
        result = fuzzy_match(lb_address, pw_address, l_id, lb_price, pw_price, lb_date, pw_date)
        if result:
            found = True
            retlist.append(result) 
            break
    if not found:
        retlist.append(None)
```

```python
third_match['check_zip'] = [int(x.split('-')[0]) if '-' in x else int(x) for x in third_match['postal_code']]

retlist = []
for lb_address, lb_price, lb_date, lb_zip in tqdm(zip(off_market_sales['SITE_ADDR'], off_market_sales['VAL_TRANSFER'], off_market_sales['DATE_TRANSFER'], off_market_sales['SITE_ZIP']), total=len(off_market_sales)):
    check = third_match[third_match['check_zip'] == lb_zip]
    found = False
    for pw_address, pw_price, pw_date, l_id in zip(check['normalized_address'], check['close_price'], check['close_date'], check['listing_id']):
        result = fuzzy_match(lb_address, pw_address, l_id, lb_price, pw_price, lb_date, pw_date)
        if result:
            found = True
            retlist.append(result)
            break
    if not found:
        retlist.append(None)
```
