```python
retdict = {}
for lb, lkup in tqdm(
    zip(lookup_check["Lightbox Field"], lookup_check["Lookup Codes"]),
    total=len(lookup_check),
):
    if (lb != "suffix_type") and (lb != "bf_stories_number"):
        tmp = (
            df.groupby(lb)["ASSESSMENT_LID"]
            .count()
            .to_frame(name="deed_count")
            .reset_index()
        )
        tmp[lb] = tmp[lb].astype(str)
        tmp = tmp.merge(
            assessment_lookups[assessment_lookups["Field Name"] == lkup],
            how="left",
            left_on=lb,
            right_on="Code",
        )
        retdict[lb] = tmp
```

```python
retdf = pd.DataFrame()
for i in retdict:
    tmp = retdict[i]
    tmp["Lightbox Field Name"] = [tmp.columns[0]] * len(tmp)
    tmp["% populated"] = tmp["deed_count"] / len(df)
    tmp.columns = ["Lightbox Field Value"] + list(tmp.columns[1:])
    # ['Lightbox Field Value', 'deed_count', 'Lookup Field Name', 'Code', 'Code Description', 'Display Code Description', 'Lightbox Field Name', '% populated']
    retdf = pd.concat(
        [
            retdf,
            tmp[
                [
                    "Lightbox Field Name",
                    "Lightbox Field Value",
                    "deed_count",
                    "% populated",
                    "Field Name",
                    "Code Description",
                ]
            ],
        ]
    )

retdf.head()
```

```python
def report_writer(dfs:dict, report_name:str):
    """Creates an excel report using a dictionary of dfs.

    Args:
        dfs (dict): Data
        report_name (str): Name of file.
    """
    writer = pd.ExcelWriter(report_name, engine='xlsxwriter')
    workbook = writer.book
    format = workbook.add_format()
    format.set_align('center')
    for sheetname, df in dfs.items():
        df.to_excel(writer, sheet_name=sheetname, index=False)
        worksheet = writer.sheets[sheetname]
        for idx, col in enumerate(df):
            series =df[col]
            max_len = max((
                series.astype(str).map(len).max(),
                len(str(series.name))
            ))+1
            worksheet.set_column(idx, idx, max_len, format)

    writer.close()
```

```python
from geopandas import gpd
import folium
import shapely
import pandas as pd
```

```python
def load_geos(fname):
    df = pd.read_csv(fname)
    df.columns = [x.upper() for x in df.columns]
    df['GEOMETRY'] = [shapely.from_wkt(x) for x in df['GEOMETRY']]
    return gpd.GeoDataFrame(df['GEOMETRY'],geometry='GEOMETRY', crs='EPSG:4326')
```

```python
geos = load_geos('data/SF_Professional_OH_20240715/Buildings_OH.csv')
```

```python
m = folium.Map(location=(41.955163, -93.494805), tiles="CartoDB positron", zoom_start=6, control_scale=True)
buildings_viz = folium.GeoJson(geos.sample(frac=.1))
buildings_viz.add_to(m)
m.save('data/building_geos.html')
```
