```python
def load_lightbox_data(fpath: str, service_zips: list) -> pd.DataFrame:
    df = pd.read_csv(fpath)
    df = df[
        [
            "DEED_LID",
            "ASSESSMENT_LID",
            "DATE_TRANSFER",
            "SITE_ADDR",
            "SITE_HOUSE_NUMBER",
            "SITE_DIRECTION",
            "SITE_STREET_NAME",
            "SITE_MODE",
            "SITE_UNIT_NUMBER",
            "SITE_CITY",
            "SITE_STATE",
            "SITE_ZIP",
            "SITE_PLUS_4",
            "DATE_FILING",
            "VAL_TRANSFER",
            "LONGITUDE",
            "LATITUDE",
        ]
    ]
    print(df['SITE_ZIP'])
    df = df[df["SITE_ZIP"].isin(service_zips)]
    df.dropna(subset="SITE_ADDR", inplace=True)
    df.dropna(subset="VAL_TRANSFER", inplace=True)
    return df[~(df["SITE_ADDR"] == "N A")]
```

```python
service_zips = pd.read_csv("data/cincy_service_zips_2024_08_28.csv").dropna()
df = load_lightbox_data(
        f"data/SF_HistoricalTransaction_OH_20240731/Deed_OH.csv", service_zips['zip']
    )
```
