```python
mapping = pd.read_csv('data/Updated Cincy Lightbox Deed Mapping - Mapping.csv')
mapping = mapping[mapping['Perchwell Attribute']!='-']
lookup_check = mapping[~pd.isna(mapping['Lookup Name'])]
lookup_check.head()
```

```python
lookups = pd.read_csv('data/Updated Cincy Lightbox Deed Mapping - Code Mappings.csv')#, dtype={'Code': str})
deeds_lookups = lookups[(lookups['Data Type (A=Assessor, D=Deed, S=Mortgage, N=NOD)'].str.contains('D')) | (lookups['Field Name']=='Standardized Land Use')][['Field Name', 'Code', 'Code Description', 'Display Code Description']]
#deeds_lookups['Code Description'] = lookups[lookups['Data Type (A=Assessor, D=Deed, S=Mortgage, N=NOD)'].str.contains('D')]['Display Code Description'].combine_first(lookups[lookups['Data Type (A=Assessor, D=Deed, S=Mortgage, N=NOD)'].str.contains('D')]['Code Description'])
deeds_lookups.head()
```

```python
retdict = {}
for lb, lkup in tqdm(zip(lookup_check['Lightbox Field'], lookup_check['Lookup Name']), total=len(lookup_check)):
    tmp = deeds_df.groupby(lb)['DEED_LID'].count().to_frame(name='deed_count').reset_index()
    tmp[lb] = tmp[lb].astype(str)
    tmp = tmp.merge(deeds_lookups[deeds_lookups['Field Name'] == lkup], how='left', left_on=lb, right_on='Code')
    retdict[lb] = tmp
    #print(tmp[pd.isna(tmp['Code Description'])])
```

```python
retdf = pd.DataFrame()
for i in retdict:
    tmp = retdict[i]
    tmp['Lightbox Field Name'] = [tmp.columns[0]]*len(tmp)
    tmp['% populated'] = tmp['deed_count']/len(deeds_df)
    tmp.columns = ['Lightbox Field Value', 'deed_count', 'Lookup Field Name', 'Code', 'Code Description', 'Display Code Description', 'Lightbox Field Name', '% populated']
    retdf = pd.concat([retdf, tmp[['Lightbox Field Name','Lightbox Field Value', 'deed_count', '% populated', 'Lookup Field Name', 'Code Description', 'Display Code Description']]])

retdf.head()
```

```python
def report_writer(dfs:dict, report_name:str):
    """Creates an excel report using a dictionary of dfs.

    Args:
        dfs (dict): Data
        report_name (str): Name of file.
    """
    writer = pd.ExcelWriter(report_name, engine='xlsxwriter')
    workbook = writer.book
    format = workbook.add_format()
    format.set_align('center')
    for sheetname, df in dfs.items():
        df.to_excel(writer, sheet_name=sheetname, index=False)
        worksheet = writer.sheets[sheetname]
        for idx, col in enumerate(df):
            series =df[col]
            max_len = max((
                series.astype(str).map(len).max(),
                len(str(series.name))
            ))+1
            worksheet.set_column(idx, idx, max_len, format)

    writer.close()
```
