## Data Ingestion
This notebook reads in the data provided by DE and cleans it.
The input files are:
- `crmls_units.csv`
- `crmls_potential_unit_buildings.csv`

```python
import pandas as pd
from ast import literal_eval
pd.set_option("display.max_columns", None)
```

# crmls_units

```python
# crmls_units
cr_units_raw = \
    pd.read_csv("data/crmls_units.csv",dtype=str,escapechar="\\")
# cr_units_raw = cr_units_raw.copy()
# cr_units_raw = cr_units_raw.sample(frac=.05)
# cr_units_raw['Assessments'] = cr_units_raw['Assessments'].fillna('{}')
cr_units_raw['Assessments'] = cr_units_raw['Assessments'].apply(literal_eval)
tempdf = pd.DataFrame(cr_units_raw['Assessments'].values.tolist(),index=cr_units_raw.index, columns=["test"])
tempdf2 = tempdf['test'].apply(pd.Series)
cr_units = pd.concat([cr_units_raw, tempdf2],axis=1) #, ignore_index=True

```

## crmls_potential_unit_buildings

```python
cr_pot_unit_blds_raw = \
    pd.read_csv("data/crmls_potential_unit_buildings.csv",escapechar="\\")
# cr_pot_unit_blds_raw = cr_pot_unit_blds_raw.copy()
# cr_pot_unit_blds_raw = cr_pot_unit_blds_raw.sample(frac=.05)
cr_pot_unit_blds_raw['Assessments'] = cr_pot_unit_blds_raw['Assessments'].apply(literal_eval)
tempdf = pd.DataFrame(cr_pot_unit_blds_raw['Assessments'].values.tolist(),index=cr_pot_unit_blds_raw.index, columns=["test"])
tempdf2 = tempdf['test'].apply(pd.Series)
cr_pot_unit_blds = pd.concat([cr_pot_unit_blds_raw, tempdf2],axis=1)
```

```python
del tempdf,tempdf2, cr_units_raw,cr_pot_unit_blds_raw
```

```python
cr_units.to_csv("clean_crmls_units.csv",index=False)
```

```python
cr_pot_unit_blds.to_csv("clean_crmls_potential_unit_buildings.csv",index=False)
```
